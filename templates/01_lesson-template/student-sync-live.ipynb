## Markdown Cells



Markdown is a markup language (kinda like HTML) but which makes formatting really easy. Long lines in a Markdown cell will wrap automatically, however if your first line is long, JupyterLab will hide the cell editing buttons that are visible in the top right corner. Because blank lines don't actually add vertical space to the formatted text, however, just start writing on line 2 to keep these buttons visible!

### Basic Markdown Formatting


Some basic Markdown syntax that you'll likely use a lot is demonstrated below:

#### Text Emphasis



You can use Markdown to _add emphasis_, **write in boldface**, or ***even do both simultaneously*** to make certain text pop. You can also `make the text monospace` for in-line code formatting.

#### Block Quotes



>You can make block quotes to drive home a point...


>Or you can add attribution to someone. Just make sure there's at least one blank line between each quote block to force Markdown to separate them! Otherwise, they'll be smushed together.
>~ ---Somebody McWhatstheirname

#### Typesetting Mathematics using $\LaTeX$


In addition to formatting text, JupyterLab can typeset math using $\LaTeX$. To add math in-line, surround your LaTeX commands in single-dollar-signs (`$...$`), but wrap in double-dollar-signs for display-style:
$$y = mx + b$$

It is also possible to use the `align` and `equation` environments, but don't push your luck beyond those.

#### Tables


Markdown tables look like this:

| Default Column | Left-Justified Column | Centered Column | Right-Justified Column | Etc. |
|----------------|:----------------------|:---------------:|-----------------------:|-----:|
| Something      | that you want to      | include         | in a table             | row. |
| Can also       | include math inside   | with            | $\LaTeX$ in a cell     | too! |

#### Code Blocks


Basic Markdown also allows you to format multi-line blocks of code, defined between matching lines of backtics:
```markdown
    ```language-name
    code code code
    some more code
    ...
    and even more code
    ```
```
which can also use syntax highlighting if the language is specified, like the below for Python:
```python

## Designing Your Lesson


There are _lots_ of different frameworks and strategies for designing the structure of your lesson. Despite this diversity, a single theme unifies all successful lesson design strategies: active learning. Including some written instructional narrative is necessary, however they should spend way more time _doing the coding_ in the lesson than reading about it beforehand. Plus, their attention spans are short, so the sooner you prompt your students to apply something you have introduced, the better the information will stick!

### Recommended Structure for Student Coding Activity


When prompting student participation in the lesson, the following is a good four-notebook-cell structure:
1. Prompt (markdown cell)
    * Write the prompt as a MyST callout/exercise block as a clear call to action
    * Make sure the instructions are clear! Unless thinking of the algorithm is the point, the instructions should almost read like pseudocode.
2. "Blank" student code entry (code cell)
    * Line 1: `# YOUR TURN: ...` comment to make it clear that they're being prompted to act
    * Line 2: Blank line to focus student attention to where they'll enter their code
3. Completed code key (code cell)
    * Line 1: `# YOUR TURN: ...` comment identical to student entry cell
    * Rest of the cell should contain the instructor's code key that completes the prompted task
4. Student Learning Check (markdown cell)
    * Provide some way for students to check their work!
    * Ideas:
        - JupyterQuiz question to check correctness of their output
        - JupyterQuiz question to prompt & store a reflection statement
        - A MyST callout block of class `dropdown` containing some amount of the code key and/or expected output (collapsed by default)
        - A MyST solution block tied back to exercise prompt (still good idea to declare this as a collapsed dropdown)
