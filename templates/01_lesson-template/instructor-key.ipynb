### Thermochemistry
:::{exercise} Heat Transfer & Calorimetry $\vert$ [OpenStax $\S$5.2 ↗️](https://openstax.org/books/chemistry-2e/pages/5-2-calorimetry)
:label:ex:calorimetry
A 360.0-g piece of rebar (a steel rod used for reinforcing concrete) is dropped into 425 mL of water (density 1.00 g/mL) at 24.0°C. The final temperature of the water was measured as 42.7°C. Calculate the initial temperature of the piece of rebar. Assume the specific heat of steel is approximately the same as that for iron (Table 5.1), and that all heat transfer occurs between the rebar and the water (there is no heat exchange with the surroundings).
:::

````python
# YOUR TURN: Solve the exercise above here!
````

````python
# YOUR TURN: Solve the exercise above here!
## Declaring variables
var('q_r m_r C_r T_r', real=True) # Rebar properties
var('q_w m_w C_w T_w', real=True) # Water properties
var('T_f', real=True) # Common final temperature

## Declaring numerical values for variables
m_r_val = 360.0 # in grams
C_r_val = 0.449 # in J / g / degreesC

m_w_val = 425.0 # in grams
C_w_val = 4.184 # in J / g / degreesC
T_w_val = 24.0 # in degreesC

T_f_val = 42.7 # in degreesC

## Define general equations to use
E_conserv =@ q_r + q_w = 0 # Energy conservation
rebar_heat =@ q_r = m_r*C_r*(T_f - T_r) # Heat change from perspective of rebar
water_heat =@ q_w = m_w*C_w*(T_f - T_w) # Heat change from perspective of water

## Substitute water_heat & rebar_heat expressions into E_conserv to make problem equation
problem = E_conserv.subs({q_r:rebar_heat.rhs, q_w:water_heat.rhs})

## Substitute numerical values into problem equation
problem = problem.subs({m_r: m_r_val,
                        C_r: C_r_val,
                        m_w: m_w_val,
                        C_w: C_w_val,
                        T_w: T_w_val,
                        T_f: T_f_val
                       })

## Rearrange problem eqn for initial rebar temperature (will be in degreesC)
solve(problem, T_r)[0].evalf()
````

### Atomic Structure
:::{exercise} Energy & Wavelengths of Bohr-Orbit Transitions $\vert$ [OpenStax $\S$6.2 ↗️](https://openstax.org/books/chemistry-2e/pages/6-2-the-bohr-model)
:label:ex:bohr-model
What is the energy (in joules) and the wavelength (in meters) of the line in the spectrum of hydrogen that represents the movement of an electron from Bohr orbit with $n = 4$ to the orbit with $n = 6$? In what part of the electromagnetic spectrum do we find this radiation?
:::

````python
# YOUR TURN: Solve the exercise above here!
````

````python
# YOUR TURN: Solve the exercise above here!
## Bohr formula setup
### Variables
var('n_f n_i', integer=True, nonzero=True, postivie=True)
var('dE R_H', real=True)

### Constants & variable values
R_H_val = 2.179E-18
n_f_val = 6
n_i_val = 4

### Equation definition
bohr =@ dE = -R_H*((1/n_f) - (1/n_i))

## Planck & light wave formulas setup 
### Variables
var('E_photon h c lamda nu')

### Constants & variable values
h_val = 6.626E-34
c_val = 2.9979E8

### Equation definitions
lightwave =@ c = lamda*nu
planck =@ E_photon = h*nu

## Problem Solution
### Excitation energy
dE_val = float(bohr.subs({R_H: R_H_val, n_f: n_f_val, n_i: n_i_val}).rhs)
print(f'dE = {dE_val} J')

### Excitation wavelength
tosolve = planck.subs({nu: (lightwave/lamda).lhs, E_photon: dE})
wavelength = solve(tosolve, lamda)[0].subs({dE: dE_val, c: c_val, h: h_val}).evalf()
lamda_val = float(wavelength.rhs)
print(f"Excitation wavelength = {lamda_val} m (near IR region)")
````

### Colligative Properties

:::{exercise} Vapor Pressure of an Ideal Solution $\vert$ [OpenStax $\S$11.4 ↗️](https://openstax.org/books/chemistry-2e/pages/11-4-colligative-properties)
:label:ex:colligative-prop
Compute the vapor pressure of an ideal solution containing 92.1 g of glycerin, C$_3$H$_5$(OH)$_3$, and 184.4 g of ethanol, C$_2$H$_5$OH, at 40°C. The vapor pressure of pure ethanol is 0.178 atm at 40°C. Glycerin is essentially nonvolatile at this temperature. 
:::

````python
# YOUR TURN: Solve the exercise above here!
````

````python
# YOUR TURN: Solve the exercise above here!
## Define variables for mass, MW, & vapor pressure for each solution component
gly_mass = 92.1    # g
gly_MW = 92.094    # g/mol
Pstar_gly = 0      # atm
etoh_mass = 184.4  # g
etoh_MW = 46.069   # g/mol
Pstar_etoh = 0.178 # atm

## Convert masses of glycerin & ethanol to moles
gly_mol = gly_mass * (1/gly_MW)
etoh_mol = etoh_mass * (1/etoh_MW)

## Compute mole fractions of ethanol and glycerin
X_etoh = etoh_mol / (etoh_mol + gly_mol)
X_gly = gly_mol / (etoh_mol + gly_mol)

## Compute solution vapor pressure from ethanol vapor pressure from Raoult's Law
P_soln = Pstar_gly * X_gly + Pstar_etoh * X_etoh

print(f"Vapor pressure of ideal solution at 40C: P = {P_soln:4.3f} atm")
````

The reaction of butadiene gas (C$_4$H$_6$) to yield C$_8$H$_{12}$ gas is described by the equation:

This “dimerization” reaction is second order with a rate constant equal to $5.76\times 10^{−2}\ {\rm L\ mol^{-1} min^{−1}}$ under certain conditions. If the initial concentration of butadiene is 0.200 M, what is the concentration after 10.0 min?

:::{hint}👀 Hint

For the general second order elementary reaction 
$$2 {\rm A}\; \rightleftarrows\; {\rm products}$$
the 2nd order integrated rate law (taken from [Table 12.2 in OpenStax $\S$12.4 ↗️](https://openstax.org/books/chemistry-2e/pages/12-4-integrated-rate-laws#fs-idm117482272)) is given by
$$\frac{1}{[A]} = kt + \frac{1}{[A]_0}$$
::::::

````python
# YOUR TURN: Solve the exercise above here!
````

````python
# YOUR TURN: Solve the exercise above here!
## Variable declaration
var('A k t A_0', real=True, nonnegative=True)

## Constants given in problem
k_val = 5.76E-2 # M/min
A_0_val = 0.200 # M
t_val = 10 # min

## Define named SymPy expression for 2nd order integrated rate law
integrated_2nd_order =@ (1 / A) = k*t + (1 / A_0)

## Solve for the concentration of A at later time, and float evaluate
solve(integrated_2nd_order.subs({k: k_val, A_0: A_0_val, t: t_val}), A)[0].evalf()
````

### Chemical Equilibrium

:::{exercise} Equilibrium Concentrations from Initial Concentrations $\vert$ [OpenStax $\S$13.4 ↗️](https://openstax.org/books/chemistry-2e/pages/13-4-equilibrium-calculations)
:label:ex:eqm
Acetic acid (CH$_3$CO$_2$H) reacts with ethanol (C$_2$H$_5$OH) to form water and ethyl acetate (CH$_3$CO$_2$C$_2$H$_5$) in a solution where dioxane (C$_4$H$_8$O$_2$; d) is the solvent:
$${\rm CH_3CO_2H\ (d) + C_2H_5OH\ (d)} \rightleftharpoons {\rm CH_3CO_2C_2H_5\ (d) + H_2O\ (d)}$$

The equilibrium constant for this reaction in dioxane is $K_c = 4.0$. What are the equilibrium concentrations for a mixture that is initially 0.15 M in CH$_3$CO$_2$H, 0.15 M in C$_2$H$_5$OH, 0.40 M in CH$_3$CO$_2$C$_2$H$_5$, and 0.40 M in H$_2$O?

:::{hint}👀 Hints
:icon:false
1. This macroscopic equilibrium is occurring in the solvent dioxane --- as indicated in the reaction by the parenthetical phases "(d)" for all species --- _not aqueous solution._ Because it's not an aqueous system, don't forget to include water in your ICE table and reaction quotient expression!!
2. Don't forget to check which direction the reaction will proceed towards equilibrium by comparing the initial value of $Q$ to the equilibrium constant $K_c$.
3. Rather than trying to solve the entire problem in a single code cell, break it into separate cells where convenient in the problem solving process.
::::::

````python
# YOUR TURN: Solve the exercise above here!
## Define Python variables for formal concentrations (F) of each species & eq'm constant (Kc)
Kc = 4.0
F_HAc = 0.15  # M
F_EtOH = 0.15 # M
F_EtAc = 0.40 # M
F_H2O = 0.40  # M

## Define SymPy variables & reaction quotient expression
### Reaction quotient Q & eq'm constant K; define scopes to be positive & real
var('Q', real=True, positive=True)
var('K', real=True, positive=True)
### Change in formal concentration, x; define scope to be real
var('x', real=True)
### Concentrations of each species; define scope to be positive & real
HAc = Symbol('[HAc]')
EtOH = Symbol('[EtOH]')
EtAc = Symbol('[EtAC]')
H2O = Symbol('[H2O]')

### Define SymPy expression for reaction quotient, assign to Python variable rxn_quotient
rxn_quotient =@ Q = (EtAc * H2O) / (HAc * EtOH)

## Evaluate initial reaction quotient using formal concentrations
Q_i = rxn_quotient.subs({EtOH: F_EtOH, HAc: F_HAc, EtAc: F_EtAc, H2O: F_H2O}).evalf().rhs

## Compare initial Q value to K, determine direction of net reaction towards equilibrium
print("Q < K...net rxn rightwards!") if bool(Q_i < Kc) else print("Q > K...net rxn leftwards!")

## Set up reaction quotient expression at equilibrium, with net rxn in correct direction
K_expression = rxn_quotient.subs({Q: Kc,
                                  HAc: F_HAc + x,
                                  EtOH: F_EtOH + x,
                                  EtAc: F_EtAc - x,
                                  H2O: F_H2O - x
                                 })

## Solve K expression for unknown change x
x1_expr, x2_expr = solve(K_expression, x)
x1 = float(x1_expr.rhs)
x2 = float(x2_expr.rhs)
print(f"Solutions to K expression: x1 = {x1}, x2 = {x2}")
print(f"Rejecting solution x1 = {x1}, keeping solution x2 = {x2}")
x_val = x2

## Compute final equilibrium concentrations
E_HAc = F_HAc + x_val  # M
E_EtOH = F_EtOH + x_val # M
E_EtAc = F_EtAc - x_val # M
E_H2O = F_H2O - x_val  # M
print("\nEquilibrium concentrations:")
print(f"\t[HAc]eq = {E_HAc:4.3f} M")
print(f"\t[EtOH]eq = {E_EtOH:4.3f} M")
print(f"\t[EtAc]eq = {E_EtAc:4.3f} M")
print(f"\t[H2O]eq = {E_H2O:4.3f} M")

## Check that equilibrium concentrations get close to equilibrium constant when plugged back into Q expr
sanity = bool(rxn_quotient.subs({Q: Kc,
                                 HAc: E_HAc,
                                 EtOH: E_EtOH,
                                 EtAc: E_EtAc,
                                 H2O: E_H2O
                                 })
             )
print(f"\nSanity Check: Does plugging eq'm []s back into Q recover K? {sanity}")
````

## Auto-Generation of Student Versions


To maximize the usability of these educational materials for adopting instructors, each lesson should be provided in three different versions:

| Modality     | Pedagogy                | Role of Instructor                                                       | Lesson Version            |
|:------------:|:-----------------------:|:------------------------------------------------------------------------:|:-------------------------:|
| Asynchronous | Guided Inquiry Learning | N/A (independent student learning)                                       | `student-async.ipynb`     |
| Synchronous  | Guided Inquiry Learning | Facilitate student process, answer questions, ensure nobody falls behind | `student-sync-gil.ipynb`  |
| Synchronous  | LiveCoding              | Demonstrate & narrate process, control pacing, etc.                      | `student-sync-live.ipynb` |


Managing these student versions is a significant burden for the author during both the initial development of the lesson and in any subsequent phases of its revision. Fortunately, it is possible to use the `jupyter-nbconvert` package to _automatically_ generate the three student versions.[^fn:nbconvert-student-versions]

To use `jupyter-nbconvert` to automatically generate the student versions of this lesson,
1. Define which cells should be deleted in each student version by adding cell tags to each cell in this notebook
    * See table below for which cell tags should be applied to each cell
    * Refer to [this section of the MySTMD documentation (launch ↗️)](https://mystmd.org/guide/notebook-configuration#notebook-cell-tags) for a step-by-step tutorial
2. Execute each of the command cells below to generate the student lesson versions

[^fn:nbconvert-student-versions]: Thanks to Steve Singleton for figuring out how to use `jupyter-nbconvert` to do this!
