:::{tip}🐍 Declaring SymPy Variables

:::{note}ℹ️ Additional Notes
:icon:false

2. By default, SymPy variables are assumed to be complex numbers (i.e., `var('x')` would define the number $x = a + bi\in\mathbb{C}$). To define a variable with a different mathematical domain, pass them as additional keyword arguments to the `var()` function:
    ```{code} python
    :linenos:

    var('m n', real=True)

    var('y', real=True, positive=True, nonzero=True)
    # z is a nonnegative real number
    var('z', real=True, nonnegative=True)
    ```
    Note that these keyword arguments apply to all variables defined in the same call to `var()`, so variables with different domains must be declared separately.

Now that our math variables and constants have been declared, we can use them to define a mathematical expression. 

:::{tip}🐍 Defining SymPy Expressions
:icon:false
Mathematical expressions are defined between numbers, SymPy variables, and defined mathematical functions in SymPy using binary arithmetic operators, with syntax
```{code} python
:linenos:
:filename:Types of SymPy Expressions
# Dummy variables for illustration purposes
var('a b c x y z')

# Anonymous SymPy expression
x - 3*y + z**4

# Named SymPy expression without an equals sign, assigned to Python variable with `=`
eqn_without_equals_sign = a*x**2 + b*x + c

# Named SymPy expression containing a mathematical equals sign, assigned to Python variable with `=@`
eqn_with_equals_sign =@ a = b / c
```

This reads just like a normal Python assignment statement, but uses the `=@` symbol to assign the $a = \frac{b}{c}$ `Equation` object to the Python variable named `eqn_name` because, in general, `Equation` objects can contain an equals sign (which, recall, is the symbol used to assign a value to a Python variable).

:::{note}ℹ️ Additional Notes on SymPy Expressions
:icon:false
1. Multiplication _must be explicitly indicated_ using the `*` symbol. Otherwise, SymPy will throw an error for an undefined symbol.
   ```python
    var('x')
    2 * x # This is correct, always need to explicitly show multiplication
    2x    # Will throw an error because `2x` isn't a defined SymPy variable
    ```
2. Multiplication **is not implied by parentheses**! You _can_ use parentheses to separate terms and for order-of-operations, but the Python interpreter will assume that a number or SymPy variable immediately preceeding a left parenthesis is a function call! To define the expression $x(x-2)^2$, 
    ```python
    var('x')
    x * (x - 2)**2 # This is correct, always need to explicitly show multiplication
    x(x-2)**2      # Will throw an error because `x` isn't a defined function
    ```
3. The special assignment operator `=@` can only be used to assign a SymPy `Equation` to a Python variable when there is a "mathematical equals sign" in the expression, i.e., $x=5$. If the `Equation` does not contain a mathematical equals sign (e.g., $5x + 2$), you must use the standard Python assignment operator `=` to assign the SymPy `Equation` to a Python variable.
4. The special assignment operator `=@` is only valid syntax when using the `algebra_with_sympy` wrapper library, and is not even included in standard SymPy!
:::

## "Manually" Rearranging Equations to Solve for a Variable

Now that we have defined our ideal gas law as an `Equation` object rand assigned it to a Python variable `ideal_gas`, what if we want to rearrange the equation to solve for a particular math variable? For example, to solve the expression $c = \lambda\nu$ for $\nu$ would require us to divide both sides of the equation by $\lambda$:
$$c = \lambda\nu \Rightarrow \frac{c}{\lambda} = \frac{\lambda\nu}{\nu} \Rightarrow \nu = \frac{c}{\lambda}$$
To do this symbolically in Python, we will first need to define an Equation object for the relationship $c=\lambda\nu$.

:::{warning} ⚠️ Using $\lambda$ as a SymPy Variable
:icon:false
When written out, the Greek letter $\lambda$ is spelled "lambda" (with a "b"), however `lambda` is already a reserved word in Python with a specific syntactic meaning. Therefore to use $\lambda$ as a SymPy variable in a mathematical expression, you must spell it _without_ the "b". To see this in action, execute the cell below!
:::

````python
# EXECUTE: Using Greek letter "lambda" as a SymPy variable
var('c lamda nu')
eqn =@ c = lamda * nu
eqn
````

Now that we have defined our equation $c = \lambda\nu$ and assigned it to the Python variable `eqn`, we can algebraically rearrange the expression by performing arithmetic operations on the expression itself. 
~~~{hint} 🐍 Manual Algebra: Arithmetically Manipulating a SymPy `Equation`
:icon: false

To solve for $\nu$, we need to divide both the right-hand-side (RHS) and left-hand-side (LHS) of our equation by $\lambda$. Rather than doing this once on each side and then re-equating the two, we can perform a single arithmentic operation on the entire equation that is done to both sides simultaneously:
```{code}python
:linenos:
:filename:Cell Input
eqn / lamda
```
:::{note}⏎ Cell Output
:icon:false
$\frac{c}{\lambda} = \nu$
:::

It is also possible to daisy-chain these arithmetic operations if more than one is needed to rearrange the expression in the desired manner. For example, to rearrange the Ideal gas law to solve for the number of moles, $n$:

```{code}python
:linenos:
:filename:Cell Input
ideal_gas / R / T
```
:::{note}⏎ Cell Output
:icon:false
$\frac{PV}{RT} = n$
:::
~~~

## Evaluating an `Equation` using the `.subs()` Method

Once you've rearranged your `Equation` to solve for the desired math variable, you can evaluate your `Equation` by substituting in values for the other math variables and constants. To do this, we will construct a _Python dictionary_ associating each math variable with a value to substitute:

:::{hint} 🐍 Brief Aside: Python Dictionaries
:icon: false
Python `dict`s are objects which associate values with keys that are used to "look them up," and are defined by placing comma-separated `key: value` pairs inside curly braces like the following:
```python
new_dict = {'key_1': 'value 1', 'key_2': 'value 2', 'key_3': 3, 'key_4': ideal_T}
```
In general, a dictionary's keys should be defined as strings (type `str`), and behave like Python variables that only live inside the dictionary. The dictionary's values, on the other hand, can be any mixture of types (i.e., not all entries in the dictionary must be of the same type). To access the value associated with a particular key in the dictionary, you would use square brackets `[]`:

```{code}python
:linenos:
:filename: Cell Input
print(new_dict[key_1])
```
```{code}txt
:filename: Cell Output
value_1
```
This is similar to accessing a member of a `list`, except instead of passing the member's index, you pass the name of the desired value's key inside the brackets.

## Units 

By using the `algebra_with_sympy` wrapper library, it is also possible to use units in SymPy expressions and for the units of a particular quantity to be carried throughout the algebraic manipulation of a SymPy `Equation`. Units themselves must first be declared using the `units()` function, which behaves the same way as `var()`. 

:::{tip}🐍 Declaring Units with `units()`
:icon:false
To declare units of meters, seconds, and kilograms:
```{code}python
:linenos:
:filename:Syntax for the units() function
units('m s kg')
```
Once units have been declared, they behave equivalently to SymPy variables declared with `var()`. _both_ values and units can be substituted into an `Equation` object by passing a dictionary to the `Equation.subs()` method. 
:::

~~~{tip}🐍 Substituting Values _with_ Units into a SymPy `Equation`
:icon:false
To evaluate the pressure of 1.00 mol of an ideal gas at 273 K in a 24.0 L vessel, we can either keep track of the units in our heads or substitute unit-aware values into the `Equation` itself.
```{code}python
:linenos:
:filename:Python Input
var('P V n R T')
units('L atm mol K')
d = {R: 0.08206*L*atm/mol/K, T: 273*K, n: 1.00*mol, V: 24.0*L}
ideal_P.subs(d)
```
:::{note}⏎ Cell Output
:icon:false
$P = 0.9334325atm$
:::
~~~

:::{warning}⚠️ Warning About Units
:icon:false
Units defined using the `units()` function are unaware of any relationships between them, either by way of metric prefix, definition, or conversion factor. This represents a major difference between how we might expect declared units to behave and how they _actually_ behave. Defining and using units this way within SymPy expressions is therefore almost always more trouble than it is worth.[^fn:algwsym-units]

[^fn:algwsym-units]: This is the opinion of the lesson author, based on three years of experience using the `algebra-with-sympy` wrapper library for both his personal purposes and when teaching these skills in his Physical Chemistry courses. The only significant use case he has found for this functionality is the extremely complicated dimensional analysis that arises in the context of deriving anharmonic rovibrational spectroscopic constants from quantum chemically constructed potential energy surfaces.
:::

## Automatically Solving SymPy Equations for a Particular Variable

Let's determine the pH of a 0.250 M solution of formic acid, whose acid dissociation constant is $K_a = 1.8\times 10^{-4}$. This corresponds to the macroscopic equilibrium
$${\rm HA} + {\rm H_2O} \leftrightharpoons {\rm H_3O^+} + {\rm A^-}$$

Taking the formal concentrations of hydronium and formate to be 0 M, the ICE table for this system implies that at equilibrium, the reaction quotient expression becomes
$$ K_a = \frac{[{\rm H_3O^+}]_{\rm eq}[{\rm A^-}]_{\rm eq}}{[{\rm HA}]_{\rm eq}} = \frac{x^2}{F - x},$$
where $F = 0.250\ {\rm M}$ is the formal concentration of our formic acid solution and $x$ is the change in formal concentrations necessary for the system to reach equilibrium. While we could of course rearrange and solve using the quadratic formula, we can also use the SymPy `solve()` function to solve arbitrary expressions for a specified variable automatically!

:::{tip}🐍 Using the SymPy `solve()` Function
:icon:false
To solve a named SymPy expression containing a mathematical equals for a particular variable using the `solve()` function, use the syntax
```{code}python
:linenos:
:filename:Syntax for the SymPy solve() function
var('x')
expr =@ lhs = rhs # LHS and RHS are expressions involving x, equated with mathematical equals sign
solve(expr, x)
```
:::{danger}☢️ Danger: Beware substituting SymPy variables with units attached into `solve()`
:icon:false
Even though it is possible to do algebra with units using `algebra_with_sympy`, this functionality completely breaks when using `solve()` to solve for a SymPy variable if units are also subsituted into the equation.

For everyone's sanity, ***do not attach units to values which are substituted into an equation that will be solved with*** `solve()`.
::::::

To illustrate how to do so, let's set up the equilibrium constant expression for our formic acid equilibrium.

### Chemical Equilibrium

:::{exercise} Equilibrium Concentrations from Initial Concentrations $\vert$ [OpenStax $\S$13.4 ↗️](https://openstax.org/books/chemistry-2e/pages/13-4-equilibrium-calculations)
:label:ex:eqm

$${\rm CH_3CO_2H\ (d) + C_2H_5OH\ (d)} \rightleftharpoons {\rm CH_3CO_2C_2H_5\ (d) + H_2O\ (d)}$$

The equilibrium constant for this reaction in dioxane is $K_c = 4.0$. What are the equilibrium concentrations for a mixture that is initially 0.15 M in CH$_3$CO$_2$H, 0.15 M in C$_2$H$_5$OH, 0.40 M in CH$_3$CO$_2$C$_2$H$_5$, and 0.40 M in H$_2$O?

:::{hint}👀 Hints
:icon:false
1. This macroscopic equilibrium is occurring in the solvent dioxane --- as indicated in the reaction by the parenthetical phases "(d)" for all species --- _not aqueous solution._ Because it's not an aqueous system, don't forget to include water in your ICE table and reaction quotient expression!!
2. Don't forget to check which direction the reaction will proceed towards equilibrium by comparing the initial value of $Q$ to the equilibrium constant $K_c$.
3. Rather than trying to solve the entire problem in a single code cell, break it into separate cells where convenient in the problem solving process.
::::::

~~~{solution}ex:eqm

```{code}python
:linenos:
:filename:Python Input
# YOUR TURN: Solve the exercise above here!
## Define Python variables for formal concentrations (F) of each species & eq'm constant (Kc)
Kc = 4.0
F_HAc = 0.15  # M
F_EtOH = 0.15 # M
F_EtAc = 0.40 # M
F_H2O = 0.40  # M

## Define SymPy variables & reaction quotient expression
### Reaction quotient Q & eq'm constant K; define scopes to be positive & real
var('Q', real=True, positive=True)
var('K', real=True, positive=True)
### Change in formal concentration, x; define scope to be real
var('x', real=True)
### Concentrations of each species; define scope to be positive & real
HAc = Symbol('[HAc]')
EtOH = Symbol('[EtOH]')
EtAc = Symbol('[EtAC]')
H2O = Symbol('[H2O]')

### Define SymPy expression for reaction quotient, assign to Python variable rxn_quotient
rxn_quotient =@ Q = (EtAC * H2O) / (HAc * EtOH)

## Evaluate initial reaction quotient using formal concentrations
Q_i = rxn_quotient.subs({EtOH: F_EtOH, HAc: F_HAc, EtAC: F_EtAc, H2O: F_H2O}).evalf().rhs

## Compare initial Q value to K, determine direction of net reaction towards equilibrium
print("Q < K...net rxn rightwards!") if bool(Q_i < Kc) else print("Q > K...net rxn leftwards!")

## Set up reaction quotient expression at equilibrium, with net rxn in correct direction
K_expression = rxn_quotient.subs({Q: Kc,
                                  HAc: F_HAc + x,
                                  EtOH: F_EtOH + x,
                                  EtAc: F_EtAc - x,
                                  H2O: F_H2O - x
                                 })

## Solve K expression for unknown change x
x1_expr, x2_expr = solve(K_expression, x)
x1 = float(x1_expr.rhs)
x2 = float(x2_expr.rhs)
print(f"Solutions to K expression: x1 = {x1}, x2 = {x2}")
print(f"Rejecting solution x1 = {x1}, keeping solution x2 = {x2}")
x_val = x2

## Compute final equilibrium concentrations
E_HAc = F_HAc + x_val  # M
E_EtOH = F_EtOH + x_val # M
E_EtAc = F_EtAc - x_val # M
E_H2O = F_H2O - x_val  # M
print("\nEquilibrium concentrations:")
print(f"\t[HAc]eq = {E_HAc:4.3f} M")
print(f"\t[EtOH]eq = {E_EtOH:4.3f} M")
print(f"\t[EtAc]eq = {E_EtAc:4.3f} M")
print(f"\t[H2O]eq = {E_H2O:4.3f} M")

## Check that equilibrium concentrations get close to equilibrium constant when plugged back into Q expr
sanity = bool(rxn_quotient.subs({Q: Kc,
                                 HAc: E_HAc,
                                 EtOH: E_EtOH,
                                 EtAc: E_EtAc,
                                 H2O: E_H2O
                                 })
             )
print(f"\nSanity Check: Does plugging eq'm []s back into Q recover K? {sanity}")
```
