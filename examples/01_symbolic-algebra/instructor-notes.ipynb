#### Asynchronous Guided-Inquiry Learning (`student-async.ipynb`)

Inspired by the process-oriented guided-inquiry learning (POGIL) pedagogical framework,[^fn:pogil] which has a strong tradition of adoption for both physical chemistry laboratory[^fn:pogil-pcl] and computer science education,[^fn:cspogil] this version of the lesson is designed to guide students through the lesson independently without instructor oversight or intervention.

[^fn:pogil]: [Learn more about the POGIL project on their website! (launch ↗️)](https://www.pogil.org/)
[^fn:pogil-pcl]: [Learn more about POGIL within the physical chemistry lab on the POGIL-PCL website! (launch ↗️)](https://www.pogilpcl.org/)
[^fn:cspogil]: [Learn more about POGIL within computer science education on the CS POGIL website!(launch ↗️)](https://cspogil.org/)

##### Lesson Structure
* Complete instructional narrative in Markdown cells to guide students through lesson
* Scaffolded prompts for student programming tasks and exercises, with hints and worked solutions included as drop-downs for student verification (hidden by default)
* End-of-lesson student exercises with worked solutions provided (hidden by default)

##### Target Audience
* Best for students who are either _extremely_ self-motivated or moderately self-motivated that already have some Python programming experience beyond the lesson prereq's.
* The author has also used the asynchronous version of the lesson as a supplementary resource for students after synchronous LiveCoding instruction (using `sync-live.ipynb`).

##### Lesson Timing
* 3 hours dedicated time-on-task; recommend giving students _at least_ three days to work through asynchronously (e.g., assign on Monday, due on Friday)

##### Known Challenges
* It is **not** a good idea to assign this lesson for students to do asynchronously if they are novice programmers, even after teaching them Python basics in a synchronous, LiveCoding modality. They will absolutely encounter errors they do not know how to debug and have questions they will not know how to ask!

##### Lesson Structure

#### Synchronous Instruction with LiveCoding (`student-sync-live.ipynb`)

This version of the lesson is designed to be delivered synchronously by the course instructor using the LiveCoding pedagogical framework,[^fn:live-coding] a strategy characterized by active student-faculty collaboration in the process of learning to code rather than passive student observation of instructor demonstration. LiveCoding was developed and popularized by The Carpentries, a nonprofit organization which has been teaching coding and cyberinfrastructure skills to students, researchers, and professionals internationally since 1998,[^fn:the-carpentries] and has become the gold standard for "end-user" programming instruction in chemistry courses.[^fn:arm-paper]

[^fn:live-coding]: See the [LiveCoding Lesson (launch ↗️)](https://swcarpentry.github.io/swc-releases/2016.06/instructor-training/13-live/) as part of the [the full Software Carpentry instructor training resources (launch ↗️)](https://swcarpentry.github.io/swc-releases/2016.06/instructor-training/) for complete details.
[^fn:the-carpentries]: [Learn more about The Carpentries on their website! (launch ↗️)](https://carpentries.org/)
[^fn:arm-paper]: Fuchs, Ringer McDonald, Gautam, & Kazerouni, _J. Chem. Educ._ **101**, 8, 3085 (2024) [(launch doi: 10.1021/acs.jchemed.4c00219 ↗️)](https://doi.org/10.1021/acs.jchemed.4c00219)

##### Lesson Structure

* Minimal instructional narrative in Markdown cells, only includes headers and some text/code that are best left for copy/paste or direct execution (e.g., package imports)
* Prompts for student programming examples included (minimal scaffolding, no hints, no solutions)
* End-of-lesson student exercises included (no numerical answers or solution keys provided)

##### Target Audience

* Best for groups of students with generally homogeneous background familiarity with Python programming, even if all novices at minimal level of prereq's

##### Lesson Pace & Timing

* 3 contact hours to complete full lesson, commonly give Student Exercises as out-of-class homework (at least 36 hours to complete)
* Optimally delivered in three 1-contact-hour (50-55 minutes) blocks, with breaks in between to minimize student and instructor fatigue 

##### Known Challenges

* If all students are Python novices (minimal familiarity at level of lesson prereq's), pace may need to be slowed for up to 6 contact hours
* If students do not all have similar programming familiarity, LiveCoding can be challenging to manage.

##### Advice from the Lesson Author

* Use "flags" (post-it notes placed on the back of students' laptop screens facing the instructor) for students to indicate their status
    - Red: getting an error & need help debugging
    - Green: getting the correct output & ready to move on
    - Blue (optional): student has a non-error/debugging-related question
* Best to have at least one other facilitator beyond the lead instructor
    - Try to still have at least 12:1 student:facilitator ratio (lead instructor + 1 facilitator can handle 20-25 students)
    - Facilitator(s) can help debug student issues & answer non-error questions as they arise rather than halting the progress for the whole group 
* When several students fly a red flag, poll them all and group common errors together to maximize debugging efficiency
* If the same subset of students consistently have the same error or issue as each other, _physically group them near each other_ to make helping them all at once easier
* If a few students seem consistently ahead of the group (always flying green flag), don't let them be bored!
    - Promote them to peer facilitator (don't let them help others until they put up their green flag)
    - Tell them to work ahead on their own using the `sync-gil.ipynb` version of the lesson, still use red flag to get help when needed
    - This point generally applies to no more than 5% of students in a typical class
* If one student becomes irrevocably lost (either due to unfamiliarity/unpreparedness or another technical issue):
    - There _always_ seems to be at least one of these students, at least at some point in the lesson, no matter how small the group.
    - If you have another facilitator in the room (**not a classmate!**), don't derail the group...let them focus on helping the lost student until you are free while the others are working on an exercise.
    - Do not conscript a classmate to help get the lost student back on track, even if you have already "promoted" them to peer facilitator once before. Doing so creates an unfair power dynamic for both students, and can place significant strain on the interpersonal relationships in the class. 
    - You might need to transition the lost student to pair-programming with a classmate...if you do so, make sure to pair them with a classmate of a similar aptitude to avoid creating the same unfair power dynamic. 
    - If you transition the lost student to pair programming, take the time during the next break in the lesson to get them back on track. You may also need to work with them individually after class to do so.

### Frequently Encountered Technical Challenges

While the `algebra_with_sympy` wrapper library does clean up much of the user interface to SymPy, there are still several persistent issues that students commonly encounter --- as well as some specific to `algebra_with_sympy` --- that I briefly summarize below.

#### Usability of the Official Documentation

* The official SymPy documentation is very challenging to read.
    - The library's intended purpose seems to be one of ultimate mathematical generalizability rather than intuitiveness of use, and was written _by_ professional mathematicians _for_ professional mathematicians.
    - The level of theoretical mathematical rigor present in the SymPy documentation can make it inscrutable in places, particularly for undergrads who may be trying to troubleshoot their code.
* The official `algebra-with-sympy` documentation is easy to read, but not very thorough.
* _If the official SymPy and `algebra-with-sympy` documentation were easily intelligible for novice programmers, I wouldn't have had to write this lesson._
    >If a student is experiencing an unexpected `sympy` error that is not discussed below, and you as the instructor do not have formal mathematical training (beyond the typical required courses for an undergraduate degree in a physical science), I would strongly recommend searching for a relevant fix on Stack Overflow before turning to the official documentation.

#### Important to Declare SymPy Variable Domains

By default, every SymPy variable declared with the `var()` function is assumed to be complex-valued, making every new `Equation` also complex-valued. As a result, operations that _you_ think should be rapid and straightforward (e.g., using `solve()` or evaluating an odd integral over even bounds) might not seem so straightforward to SymPy...which could make the code very slow to run or even time out entirely.

> To avoid this issue, clearly define the mathematical domain of your SymPy variables at declaration time!

For example, let's say we want to define an expression in SymPy for a wavefunction over two spatial variables $\left(x,\ y\right)$ that are both real numbers and two quantum numbers $n,\ m$ which are both positive integers. Naively, we could just declare all four "variables" at once using
```python
var('x y n m')
```
If we wished to normalize this wavefunction by integrating, SymPy will assume by default that each of these variables are complex-valued, and the integral will become either incredibly complicated to perform or even impossible. 

Alternatively, we may declare these math variables along with their mathematical "scope" (what SymPy calls [variable _assumptions_](https://docs.sympy.org/latest/guides/assumptions.html#the-old-assumptions-system)) like the following:
```python
var('x y', real=True)
var('n m', real=True, positive=True, integer=True)
```
By doing so, any future operations performed by SymPy on these variables (and the `Equation`s they define) will also be done consistent with the desired chemical context of the problem.

#### Important to Re-Declare SymPy Variables

Because specifying the mathematical domain of a SymPy variable is so critical for expressions to behave as "expected," reusing a named variable (e.g., $x$ or $n$) beyond its original problem/context can lead to significant headaches that have nothing to do with the student's math or coding _in the new problem_. 

For example, consider the case of a student who must solve two problems in the same notebook:
- Problem 1: $m$ is a nonnegative real number representing mass
- Problem 2: $m$ as a positive integer representing the orbital magnetic quantum number for the hydrogen atom
 
The physical meaning of these two $m$'s are fundametally different, which manifests in their different mathematical domain. Unfortunately, because cells in a Jupyter notebook can be executed in any order, students may find themselves performing a SymPy operation on what they _believe_ to be positive integer $m$ but which is _currently_ referring to a nonnegative real number. A student in this situation will almost certainly see some unexpected (nominally incorrect) code behavior, _even if their actual code or symbolic math is correct_, just because they didn't consider the current state of their variable.

In the experience of the lesson author, these these SymPy-variable-domain-incompatibility bugs are almost impossible to find and correct, making them extremely frustrating and disheartening for students. 

> To avoid these bugs:
> 1. Train students to always declare the mathematical domain of their SymPy variables at declaration time
> 2. Train students to always re-declare their SymPy variables for each new problem/context
> 3. Train students to always re-execute their SymPy variable declaration cells ***in LEXICAL ORDER (top to bottom)*** whenever they switch between contexts

#### Beware SymPy vs Python Variable Clashes

A common strategy for solving chemical problems with SymPy is 
1. Define an equation which needs to be manipulated in some way (e.g., solving for a variable, differentiating with respect to something, etc.)
2. Substitute numerical values into the new form of the equation, and
3. Extract the final numerical answer.

In the formic acid pH example from the lesson, the quadratic equation constructed by the requisite ICE table involves both the weak acid equilibrium constant ($K_a$) and the formal concentration of formic acid ($F$). After defining SymPy variables for $K_a$ and $F$ to construct the equilibrium constant expression, students often wish to define Python variables to store the numerical values of these SymPy variables for later substitution. When doing so, students must be careful to _not use the same names for both Python variable and math variable_, as this will cause the name to be overwritten and their code to break!

>My preferred strategy for avoiding this issue is for the math variable to have the "correct" name (e.g., `K_a`), and the Python variable storing its numerical value to have `_val` appended to the variable name (i.e., `K_a_val`). That way, both variable safety is maintained and the naming convention helps clue in the user about the contents & scope of each variable.

#### Avoid Units in SymPy Expressions

Units defined with the `units()` function from the `algebra-with-sympy` wrapper library have the following limitations and known issues:
1. Units are unaware of any relationships between them, either by way of metric prefix, definition, or conversion factor. This represents a major difference between how we might expect declared units to behave and how they _actually_ behave.
2. Units do not always cancel properly and/or simplify fully when values with attached units are substituted into SymPy expressions.
3. Solving for a variable using the SymPy function `solve()` after values with attached units are subsituted for the other variables can throw unexpected errors or fail to simplify properly.

Based on the experience of the lesson author with repeatedly debugging students' issues with the above, ***the `units()` functionality should be avoided for the collective sanity of both students and instructor.***
> The only worthwhile use case that the lesson author has found for this functionality is when dimensional analysis is the rate-determining step for a problem. In the four years of using this library for teaching physical chemistry and for his own scientific purposes, this has only occurred once: when deriving fully rovibrational spectroscopic constants of diatomic molecules from first principles using _ab initio_ potential energy surfaces constructed via quantum chemistry.

#### Differentials Unsupported in SymPy

SymPy currently has no support for differentials (e.g., $dx$) as an independent mathematical entity. It is therefore not currently possible to leverage the full flexibility of Leibnitz notation in SymPy to solve differential equations by way of differential separation and integration, or to define expressions in terms of differentials (more details in the instructor notes for the Symbolic Calculus lesson).
