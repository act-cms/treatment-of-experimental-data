### Code Blocks

#### Markdown Code Blocks

It is possible to create a multi-line code block with standard Markdown syntax as
```{code}markdown
    ```
    print("This is a language-agnostic code block \
    that spans multiple lines.
    ```
```

It is also possible to add syntax highlighting for [{u}`one of these recognized programming languages`](https://github.com/react-syntax-highlighter/react-syntax-highlighter/blob/master/AVAILABLE_LANGUAGES_HLJS.MD) to the code block by specifying the language name immediately after the opening line of triple-backtics. For example, the following Markdown code
```{code}markdown
    ```python
    for i in range(10):
        print(f"this is the {i}th value to be printed")
    ```
```
would be formatted as
```python
for i in range(10):
    print(f"this is the {i}th value to be printed")
```

#### MyST `{code}` Directive


When using [{u}`the MyST ``{code}`` directive (refer to the documentation for details)`](https://mystmd.org/guide/code#code-blocks), the following additional formatting is available:
* Add a label for cross-referencing
* Add a caption to the code listing
* Add line numbers
* Highlight particular lines of code in the listing
* Add a filename to annotate the code


In Dom's experience, the two most useful of these additional formatting capabilities are filenames and line numbers, particulary when providing a glimpse at both code input and output:

```{code}python

## MyST Callouts


Beyond the additional typographical features above, one of the largest advantages of MyST over basic Markdown are _callouts_, which produce admonitions as colored boxes that set their contents apart from regular text. For full discussion of this functionality, refer to the [{u}`Callouts & Admonitions guide in the MyST Documentation`](https://mystmd.org/guide/admonitions).

### Nested Columns & Tab Sets

~~~~~~{tip}🐍 Defining a Python Function
:icon:false
Click through the tabs in the right-hand column to dissect the syntax for declaring a custom Python function.
~~~~~{grid} 2
~~~~{grid-item}
(a) Function declaration & definition syntax example
```{code}python
:linenos:
:filename: Function definition syntax
def function_name(arg1, arg2, kwarg1="default_value"):
    """This is a docstring which documents the purpose
    & behavior of the function!"""
    # Comments can be used to provide context
    pass # placeholder statement
```
~~~~

~~~~{grid-item}
(b) Breakdown of each line of code
~~~{tab-set}
::::{tab-item} Line 1: Function Declaration
Line 1 declares the named function & defines its signature.
:::{note} The `def` keyword
:icon:false
:class:dropdown
:open:false
The line must begin with the Python `def` keyword, followed by a space.
:::
:::{note} Function Name
:icon:false
:class:dropdown
:open:false
Following the space after the `def` keyword is the name of the function, which must follow the same rules as Python variable names
:::
:::{note} Parenthetical List of Function Arguments
:icon:false
:class:dropdown
:open:false
* Parentheses immediately follow the name of the function, containing comma-separated list of function arguments
* Parentheses are included even if there are no function arguments, e.g., `def foo():`
* Shouldn't have a space between function name and parentheses, but having a space won't produce a syntax error
:::
:::{note} Args vs. Kwargs
:icon:false
:class:dropdown
:open:false
Two types of arguments are possible for a Python function:
* _Positional_ arguments (a.k.a. "args") are ones without a default value, and must be listed first
* _Keyword_ arguments (a.k.a. "kwargs") are ones with a default value
    - Default value for a kwarg is defined using the assignment operator `=`
    - **Kwargs must be listed _after_ the positional arguments**, otherwise it's a syntax error!
:::
:::{note} Declaration-Ending Colon
:icon:false
:class:dropdown
:open:false
A colon `:` must immediately follow the parenthetical list of arguments to end the function declaration.
:::
::::
::::{tab-item} Lines 2-5: Function Block
Lines 2-5 define the indented _function block_ containing the code to execute.
:::{note} Function block indentation & execution
:icon:false
:class:dropdown
:open:false
* When called, the function executes the lines contained in the indented _function block_ in sequential order
* All lines of code in the function block _must be indented_ (typically using {kbd}`Tab`), otherwise it's a syntax error!
* Function blocks _must_ contain at least one line of executable clode, otherwise it's a syntax error!
* Once all lines in the function block are executed, the interpreter leaves the function and goes back to the caller
:::
:::{note} Lines 2-3: Function Docstring
:icon:false
:class:dropdown
:open:false
All functions _should_ include a "docstring", appearing on the first line of the function block
* Docstrings are defined using triple-quotes (`"""docstring contents"""`) and are of type `str`
* Unlike a `str` defined with single quotes (`'blah'` or `"yadda"`), triple-quoted strings can break across multiple lines without throwing a syntax error
* The built-in `help()` function returns the function docstring when calling `help(function_name)`
:::
:::{note} Line 4: Comment adding context to the function block itself
:icon:false
:class:dropdown
:open:false
Code comments are not returned by `help()`, so while they are helpful when looking at the function itself, are not actually visible to the caller of the function
:::
:::{note} Line 5: `pass` Statement
:icon:false
:class:dropdown
:open:false
* The `pass` keyword is a placeholder statement, which tells the interpreter to move onto the next line of code
* This line fulfills the requirement that all function blocks have at least one line of executable code
:::
::::
~~~
~~~~
~~~~~
~~~~~~

### Nested Dropdowns

::::{tip}🐍 Type Casting with Built-In Python Functions
:icon:false
To cast an existing Python object `my_obj` to a new type, use the built-in Python functions below:
* `str(my_obj)`: Cast `my_obj` to be of type `str` [(go to docs ↗️)](https://docs.python.org/3/library/functions.html#func-str)
* `int(my_obj)`: Cast `my_obj` to be of type `int` [(go to docs ↗️)](https://docs.python.org/3/library/functions.html#int)
* `float(my_obj)`: Cast `my_obj` to be of type `float` [(go to docs ↗️)](https://docs.python.org/3/library/functions.html#float)
* `bool(my_obj)`: Cast `my_obj` to be of type `bool` [(go to docs ↗️)](https://docs.python.org/3/library/functions.html#bool)

~~~~{warning}⚠️ Caution: Not all typecasts are possible or reasonable!
:icon:false
Below are three of the most common typecasting errors to beware.

:::{error}⛔️ Forbidden Typecasts
:icon:false
:class:dropdown
:open:false
Some typecasts are expressly forbidden:
```{code}python
:filename:An illegal typecast: str->float
float("Sphynx of black quartz, judge my vow!")
```
